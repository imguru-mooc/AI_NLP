# Llama 3 모델 불러오기

```python
import os
os.environ['HF_TOKEN']="여러분의 HuggingFace Access Token값"
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

```python
def generate_response(system_message, user_message):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": user_message},
    ]

    input_ids = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt"
    ).to(model.device)

    terminators = [
        tokenizer.eos_token_id,
        tokenizer.convert_tokens_to_ids("<|eot_id|>")
    ]

    outputs = model.generate(
        input_ids,
        max_new_tokens=256,
        eos_token_id=terminators,
        do_sample=True,
        temperature=0.6,
        top_p=0.9
    )
    response = outputs[0][input_ids.shape[-1]:]

    return tokenizer.decode(response, skip_special_tokens=True)
```

```python
llama3_translsation_text = generate_response(system_message="너는 번역을 수행하는 챗봇이야. 다음 내용을 한국어로 번역해줘.",
                             user_message=영어_text)
print(llama3_translsation_text)
```

```python
def print_beautifully(english, korean, llama, gpt4, gpt35, clova, gemini):
    """
    Prints the translations of a given text by different AI models in a beautifully formatted way.

    Args:
    - english (str): English original text.
    - korean (str): Korean original text.
    - llama (str): Translation by Llama.
    - gpt4 (str): Translation by GPT-4.
    - gpt35 (str): Translation by GPT-3.5.
    - clova (str): Translation by Clova AI.
    - gemini (str) : Translation by Gemini

    """
    print("### Translations Comparison ###\n")
    print(f"Original (English):\n{english}\n")
    print(f"Original (Korean):\n{korean}\n")
    print(f"Llama 3:\n{llama}\n")
    print(f"GPT-4:\n{gpt4}\n")
    print(f"GPT-3.5:\n{gpt35}\n")
    print(f"Clova X:\n{clova}\n")
    print(f"Gemini:\n{gemini}\n")
```

```python
# 대표 LLM별 번역 영-한 번역 성능 비교
english_text = "The airway route will gradually be expanded domestically, Gwangju-Jeju, Gwangju-Yangyang, Gwangju-Gimhae, as well as to Southeast Asia such as Japan and China."
korean_text = "하늘길 운항노선도 광주-제주 광주-양양 광주-김해 등 국내는 물론 일본과 중국 등 동남아로 점차 넓혀가기로 했다."
llama_text = llama3_translsation_text
gpt4_text = "항공 노선은 국내에서 점차 확장될 예정이며, 광주-제주, 광주-양양, 광주-김해 노선뿐만 아니라 일본, 중국과 같은 동남아시아로도 확장될 예정입니다."
gpt35_text = "항공로가 점차 확장될 것으로 예상되며, 국내선으로는 광주-제주, 광주-양양, 광주-김해 노선이 확장되고, 일본과 중국 등 동남아시아로의 항공로도 개설될 예정입니다."
clova_text = "국내선 항공 노선은 광주-제주, 광주-양양, 광주-김해 노선을 비롯하여 일본, 중국 등 동남아시아까지 점차 확대될 예정입니다."
gemini_text = "항공 노선은 국내적으로 광주-제주, 광주-양양, 광주-김해 노선을 비롯하여 일본, 중국 등 동남아시아로 점차 확대될 예정입니다."

# Call the function to display the example texts
print_beautifully(english_text, korean_text, llama_text, gpt4_text, gpt35_text, clova_text, gemini_text)
```
