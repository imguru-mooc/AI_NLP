```python
import spacy
import numpy as np

# Load the pre-trained language model
nlp = spacy.load('en_core_web_sm')

# Define the words
words = ["happy", "hello", "provide", "tomorrow"]

# Convert the words into vectors
vectors = []
for word in words:
    token = nlp(word)
    vectors.append(token.vector)

# Print the vectors
print("Vectors for the words:", words)
print(np.array(vectors))
```

```python
# Convert the words into 1D arrays
arrays = [np.array(nlp(word).vector) for word in words]
print("Arrays:")
print(arrays)
```

```python
# Convert the words into a 2D matrix
matrix = np.array([nlp(word).vector for word in words])
print("Matrix:")
print(matrix)
```

```python
# Convert the words into a 3D tensor
tensor = np.array([[nlp(word).vector] for word in words])
print("Tensor:")
print(tensor)
```
